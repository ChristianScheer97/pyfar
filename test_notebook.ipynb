```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import haiopy
from haiopy import Signal
from haiopy import Coordinates
from haiopy import Orientation
from haiopy import plot

import numpy as np
from scipy.io.wavfile import read

%matplotlib inline
%config InlineBackend.figure_format='retina'
%load_ext autoreload
%autoreload 2
```

```python
test_audio_read = read('test_rir_noise.wav')
sampling_rate = test_audio_read[0]
test_audio_raw = test_audio_read[1]

test_signal_object = Signal(data=test_audio_raw,
                    sampling_rate=sampling_rate,
                    domain='time',
                    signal_type='energy')
```

```python
ax = plot.plot_time(test_signal_object)
```

```python
ax = plot.plot_freq(test_signal_object)
```

```python
ax = plot.plot_phase(test_signal_object)
```
